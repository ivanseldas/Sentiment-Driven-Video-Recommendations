```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_avg_emotions = pd.read_csv('../docker_app/df_avg_emotions.csv', index_col=0).set_index('video_id')
df_avg_emotions.head()
```

```python
video_id = np.random.choice(df_avg_emotions.index)
video_id
```

* Average Emotion Score

```python
plt.figure(figsize=(12,8))
df_avg_emotions.loc[video_id].sort_values(ascending=False).drop('final_emotion_score')[:10].sort_values(ascending=True).plot(kind='barh')
final_score = df_avg_emotions.loc[video_id]['final_emotion_score']
plt.title(f'[{video_id}] Final Score: {final_score}')
plt.show();
```
